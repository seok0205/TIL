# YOLO 모델 활용해 이미지 객체 탐지

## YOLO(You Only Look Once)란?

- 객체 탐지 모델 중 하나, 이미지에서 객체의 위치와 종류를 동시에 예측하는 강력한 딥러닝 모델.
- 한 번의 신경망 전파만으로 객체를 탐지하기 때문에 실시간 처리가 가능할 정도로 빠름.

### 특징

- 속도 : YOLO는 전체 이미지를 한 번에 처리해 빠르게 객체 탐지 가능.
- 정확도 : 여러 객체가 있는 복잡한 이미지에서도 높은 정확도로 객체 탐지.
- 다양한 크기 : YOLO는 다양한 크기의 이미지와 객체 처리 가능.

## 이미지 객체 탐지

1. YOLOv8 설치

```shell
pip install ultralytics
```

2. 라이브러리 임포트

````python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
````

3. YOLOv8 모델 로드

````python
# YOLOv8 모델 로드 (YOLOv8s)
model = YOLO('yolov8n.pt')
````

4. 이미지 객체 탐지

````python
# 이미지 파일 경로
image_path = 'stove.jpg'

# 이미지를 모델에 입력하여 객체 탐지 수행
results = model(image_path)

# 탐지 결과 출력
result = results[0]
````

5. 결과 시각화

- 탐지된 객체 포함 이미지 시각화

````python
# 탐지된 객체들이 표시된 이미지를 가져옴
img_with_boxes = result.plot()  # result.plot()은 바운딩 박스가 포함된 이미지를 반환

# Matplotlib을 사용하여 이미지 출력
plt.imshow(cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
````
